# Linear Regression Model - Home Pricing Prediction

###  Importing Packages

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
import numpy as np
```

### Reading Data

```python
df = pd.read_excel("C:\Code3\dataset.xlsx")
df
```

```python
%matplotlib inline
```

```python
plt.scatter(df.Area, df.SalePrice, color='purple', marker='+')
plt.xlabel('Area Sq ft. ')
plt.ylabel('Sale Price $')
```

### Building Linear Regression Model

```python
reg_lin = linear_model.LinearRegression()
reg_lin.fit(df[['Area']], df.SalePrice)
```

```python
reg_lin.predict(np.array([3300]).reshape(-1, 1))
```

```python
%matplotlib inline
plt.scatter(df.Area, df.SalePrice, color='red', marker='+')
plt.xlabel('Area Sq ft. ')
plt.ylabel('Sale Price $')
plt.plot(df.Area, reg_lin.predict(df[['Area']]), color='blue')
```

```python
a = pd.read_csv("C:\\Code3\\areas.csv")
a.head()
```

```python
predict = reg_lin.predict(a)
predict
```

```python
a['SalePrice'] = predict
```
